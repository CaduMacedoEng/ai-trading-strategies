```python
# Compute the daily mean of all stocks. This will be our equal weighted benchmark
daily_mean = pd.DataFrame(total_return.loc[:,'F_1_d_returns'].groupby(level='Date').mean())
daily_mean.rename(columns={'F_1_d_returns':'SP&500'}, inplace=True)

# Convert daily returns to cumulative returns
cum_returns = pd.DataFrame((daily_mean[['SP&500']]+1).cumprod())

# Ploting the cumulative returns
cum_returns.plot()

# Customizing the plot
plt.title('Cumulative Returns Over Time', fontsize=16, fontweight='bold')
plt.xlabel('Date', fontsize=14)
plt.ylabel('Cumulative Return', fontsize=14)
plt.grid(True)
plt.xticks(rotation=45)
plt.legend(title_fontsize='13', fontsize='11')

# Display the plot
plt.show()

# Calculate the number of years in the dataset
days_of_trading = 252
number_of_years = len(daily_mean) / days_of_trading # Assuming 252 trading days in a year

ending_value = cum_returns['SP&500'].iloc[-1]
beginning_value = cum_returns['SP&500'].iloc[1]

# Compute the Compound Annual Growth Rate (CAGR)
ratio = ending_value/beginning_value
cagr = round((ratio**(1/number_of_years)-1)*100,2)

print(f'The CAGR is: {cagr}%')

# Compute the Sharpe Ratio by annualizing the daily meaen and the daily std
average_daily_return = daily_mean[['SP&500']].describe().iloc[1,:] * days_of_trading
stand_dev_daily_return = daily_mean[['SP&500']].describe().iloc[2,:] * pow(days_of_trading,1/2)

sharpe = average_daily_return/stand_dev_daily_return

print(f'Sharpe Ratio of Strategy: {round(sharpe.iloc[0],2)}')

# df_daily_mean.rename(columns={target: 'Strategy'}, inplace=True)
ann_returns = (pd.DataFrame((daily_mean[['SP&500']]+1).groupby(daily_mean.index.get_level_values(0).year).cumprod())-1)*100
calendar_returns = pd.DataFrame(ann_returns['SP&500'].groupby(daily_mean.index.get_level_values(0).year).last())

calendar_returns.plot.bar(rot=30, legend='top_left') #.opts(multi_level=False)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from empyrical import sharpe_ratio

# ========================
# ✅ TRATAMENTO DE VALORES NULOS
# ========================
# Assumindo que total_return é um DataFrame já carregado com retornos diários das ações
# Removendo valores nulos para evitar problemas nos cálculos
total_return.dropna(subset=['F_1_d_returns'], inplace=True)

# ========================
# ✅ CÁLCULO DE RETORNOS LOGARÍTMICOS
# ========================
# Calculando retornos logarítmicos em vez de retornos simples
total_return['log_returns'] = np.log(total_return['F_1_d_returns'] + 1)

# Agrupando por data e calculando a média diária
daily_mean = pd.DataFrame(total_return.groupby(level='Date')['log_returns'].mean())
daily_mean.rename(columns={'log_returns': 'SP&500'}, inplace=True)

# ========================
# ✅ CONVERSÃO DE RETORNOS DIÁRIOS EM RETORNOS CUMULATIVOS
# ========================
# Convertendo os retornos logarítmicos em retornos cumulativos
cum_returns = pd.DataFrame((daily_mean[['SP&500']].cumsum()).apply(np.exp))

# ========================
# ✅ PLOTAGEM DO GRÁFICO DE RETORNOS CUMULATIVOS
# ========================
# Plotando o gráfico de retornos cumulativos
plt.figure(figsize=(12, 6))
cum_returns.plot(ax=plt.gca())
plt.title('Cumulative Returns Over Time', fontsize=16, fontweight='bold')
plt.xlabel('Date', fontsize=14)
plt.ylabel('Cumulative Return', fontsize=14)
plt.grid(True)
plt.xticks(rotation=45)
plt.legend(title_fontsize='13', fontsize='11')
plt.axhline(y=1, color='r', linestyle='--', label='Baseline')
plt.annotate('COVID Crash', xy=('2020-03-01', 0.8), xytext=('2019-12-01', 1),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.show()

# ========================
# ✅ CÁLCULO DO CAGR (Compound Annual Growth Rate)
# ========================
# Calculando o CAGR (Taxa de Crescimento Anual Composta)
days_of_trading = 252  # Supondo 252 dias de negociação por ano
number_of_years = len(daily_mean) / days_of_trading

ending_value = cum_returns['SP&500'].iloc[-1]
beginning_value = cum_returns['SP&500'].iloc[0]

cagr = round(((ending_value / beginning_value) ** (1 / number_of_years) - 1) * 100, 2)
print(f'The CAGR is: {cagr}%')

# ========================
# ✅ CÁLCULO DO SHARPE RATIO (USANDO EMPYRICAL)
# ========================
# Usando a biblioteca empyrical para calcular o Sharpe Ratio
sharpe = sharpe_ratio(daily_mean['SP&500'])
print(f'Sharpe Ratio of Strategy: {round(sharpe, 2)}')

# ========================
# ✅ AJUSTE PARA INFLAÇÃO (OPCIONAL)
# ========================
# Ajustando os retornos para inflação (supondo uma inflação de 2% ao ano)
inflation_rate = 0.02
real_cum_returns = cum_returns / ((1 + inflation_rate) ** number_of_years)

# ========================
# ✅ CÁLCULO DOS RETORNOS ANUAIS (RETORNOS CALENDÁRIOS)
# ========================
# Calcular os retornos anuais usando resample com 'YE' (Year End)
calendar_returns = ((daily_mean[['SP&500']] + 1).resample('YE').prod() - 1) * 100

# Ajustar o índice para mostrar apenas o ano
calendar_returns.index = calendar_returns.index.year

# ========================
# ✅ PLOTAGEM DOS RETORNOS ANUAIS
# ========================
# Plotando os retornos anuais em um gráfico de barras
calendar_returns.plot(kind='bar', rot=45, figsize=(12, 6))
plt.title('Annual Returns')
plt.xlabel('Year')
plt.ylabel('Return (%)')
plt.grid(True)
plt.show()

# ========================
# ✅ ADICIONANDO MÚLTIPLOS BENCHMARKS (OPCIONAL)
# ========================
# Exemplo de como adicionar múltiplos benchmarks para comparação
# Supondo que você tenha retornos diários de outros índices
small_caps_returns = np.random.normal(0.0005, 0.02, len(daily_mean))
emerging_markets_returns = np.random.normal(0.0003, 0.03, len(daily_mean))

# Criando um DataFrame com múltiplos benchmarks
benchmarks = pd.DataFrame({
    'SP&500': daily_mean['SP&500'],
    'Small Caps': small_caps_returns,
    'Emerging Markets': emerging_markets_returns
})

# Plotando todos os benchmarks
plt.figure(figsize=(12, 6))
benchmarks.cumsum().apply(np.exp).plot(ax=plt.gca())
plt.title('Cumulative Returns Comparison')
plt.xlabel('Date')
plt.ylabel('Cumulative Return')
plt.legend(loc='best')
plt.grid(True)
plt.show()
```
